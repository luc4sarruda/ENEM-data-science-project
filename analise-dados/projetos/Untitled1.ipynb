```python
import pandas as pd
```

```python
microdadosEnem = pd.read_csv("../../microdados_enem_2023/DADOS/MICRODADOS_ENEM_2023.csv", sep = ";", encoding = 'ISO-8859-1')
```

```python
microdadosEnem.head()
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items = ['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
```

```python
microdadosEnemSelecionado.head()
```

```python
mediaEnemMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC'], as_index = False).mean()
```

```python
mediaEnemMunicipio
```

```python
mediaEnemMunicipio.to_csv("mediaEnemMunicipio.csv")
```

```python
atlasHumano = pd.read_csv("../../microdados_enem_2023/DADOS/atlas.csv", sep = ",")
```

```python
atlasHumano.head()
```

```python
enem_idhm = pd.merge(mediaEnemMunicipio, atlasHumano, how = 'left', left_on = ['CO_MUNICIPIO_ESC'], right_on = ['Codmun7'])
```

```python
enem_idhm.head()
```

```python
enem_idhm = enem_idhm.filter(items = ['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'SG_UF_ESC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'IDHM'])
```

```python
enem_idhm[0:9]
```

```python
enem_idhm.plot.scatter('NU_NOTA_MT', 'IDHM', figsize = (15, 15))
```

```python
media_redacao = enem_idhm.NU_NOTA_REDACAO.mean()
```

```python
media_idhm = enem_idhm.IDHM.mean()
```

```python
media_redacao
```

```python
media_idhm
```

```python
quadrante1 = (enem_idhm.IDHM < media_idhm) & (enem_idhm.NU_NOTA_REDACAO < media_redacao)
```

```python
quadrante2 = (enem_idhm.IDHM > media_idhm) & (enem_idhm.NU_NOTA_REDACAO < media_redacao)
```

```python
quadrante3 = (enem_idhm.IDHM > media_idhm) & (enem_idhm.NU_NOTA_REDACAO > media_redacao)
```

```python
quadrante4 = (enem_idhm.IDHM < media_idhm) & (enem_idhm.NU_NOTA_REDACAO > media_redacao)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2, 2, figsize = (20, 10), sharex = 'col', sharey = 'row')
plt.subplots_adjust(wspace = 0, hspace = 0)

minRedacao = enem_idhm.NU_NOTA_REDACAO.min()
minIDHM = enem_idhm.IDHM.min()
maxRedacao = enem_idhm.NU_NOTA_REDACAO.max()
maxIDHM = enem_idhm.IDHM.max()

enem_idhm1 = enem_idhm[quadrante1]
enem_idhm1.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=4, color = 'red', ax = ax[1, 0], xlim = (minRedacao, media_redacao), ylim = (minIDHM, media_idhm))

enem_idhm2 = enem_idhm[quadrante2]
enem_idhm2.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=4, color = 'yellow', ax = ax[0, 0], xlim = (minRedacao, media_redacao), ylim = (media_idhm, maxIDHM))

enem_idhm3 = enem_idhm[quadrante3]
enem_idhm3.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=4, color = 'green', ax = ax[0, 1], xlim = (media_redacao, maxRedacao), ylim = (media_idhm, maxIDHM))

enem_idhm4 = enem_idhm[quadrante4]
enem_idhm4.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=4, color = 'yellow', ax = ax[1, 1], xlim = (media_redacao, maxRedacao), ylim = (minIDHM, media_idhm))
```
